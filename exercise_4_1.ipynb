```python
from qiskit import *
from qiskit.visualization import plot_bloch_multivector
from qiskit.visualization import plot_histogram
from qiskit.quantum_info import Statevector
import numpy as np
```

```python
matplotlib inline
```

```python
# Two qubit quantum fourier transform
angle = np.pi*1/2
circuit = QuantumCircuit(2)
circuit.x([0])
circuit.barrier()
circuit.h(1)
circuit.cu1(angle, 0, 1)
circuit.h(0)
circuit.swap(0, 1)
circuit.barrier()

# Measurement is not relevant in the fourier basis
#circuit.measure(0, 0)
#circuit.measure(1, 1)
circuit.draw(output='mpl')
```

```python
state = Statevector.from_instruction(circuit)
plot_bloch_multivector(state)
```

```python
# Three qubit quantum fourier transform
angle = np.pi*1/2
qc = QuantumCircuit(3)
qc.x([1])
qc.barrier()
qc.h(2)
qc.cu1(angle, 2, 1)
qc.cu1(angle/2, 0, 2)
qc.h(1)
qc.cu1(angle, 1, 0)
qc.h(0)
qc.swap(0, 2)
qc.barrier()

# Measurement is not relevant in the fourier basis
#circuit.measure(0, 0)
#circuit.measure(1, 1)
qc.draw(output='mpl')
```

```python
state2 = Statevector.from_instruction(qc)
plot_bloch_multivector(state2)
```
